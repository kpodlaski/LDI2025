# Inicjalizacja bibliotek

```python
# @title
from PIL import Image
import numpy as np
from matplotlib import pyplot as plt
from cv2 import filter2D
```

# Obrazek i jego sk≈Çadowe RGB

```python
# @title
obrazek = Image.open("tecza.jpg")
mini_obrazek = obrazek.resize((int(obrazek.size[0]/3),int(obrazek.size[1]/3)))
mini_obrazek
```

```python
# @title
red_mini = np.array(mini_obrazek)
red_mini[:,:,1]=0
red_mini[:,:,2]=0

green_mini = np.array(mini_obrazek)
green_mini[:,:,0]=0
green_mini[:,:,2]=0

blue_mini = np.array(mini_obrazek)
blue_mini[:,:,0]=0
blue_mini[:,:,1]=0

gray_mini = np.array(mini_obrazek.convert('L'))
orig_mini = np.array(mini_obrazek)
```

```python
# @title
fig = plt.figure()
plt.subplot(232)
plt.imshow(mini_obrazek)
plt.subplot(233)
plt.imshow(gray_mini, cmap='gray', vmin=0, vmax=255)
plt.subplot(234)
plt.imshow(red_mini)
plt.subplot(235)
plt.imshow(green_mini)
plt.subplot(236)
plt.imshow(blue_mini)
plt.tight_layout()
```

# Wytnijmy sobie kawa≈Çeczek aby go dok≈Çadniej obejrzeƒá

```python
# @title
fragment = orig_mini[75:85,:10]
fragment_gray = gray_mini[75:85,:10]
fig = plt.figure()
plt.subplot(232)
plt.imshow(fragment)
plt.subplot(233)
plt.imshow(fragment_gray, cmap='gray', vmin=0, vmax=255)
```

```python
# @title
print("Rozmiar szarego fragmentu:", fragment_gray.shape)
print("Rozmiar kolorowego fragmentu", fragment.shape)
```

```python
# @title
fragment_gray
```

```python
# @title
fragment
```

Wybierzmy kernel, czyli zastosujmy jednƒÖ z poni≈ºszych kom√≥rek:

```python
# @title
# Wyg≈Çadzenie lokalne -- U≈õrednienie z najbli≈ºszymi sƒÖsiadami
kernel_base = np.array(
    [[1,1,1],
     [1,1,1],
     [1,1,1]])
kernel_base
```

```python
# @title
# Wyg≈Çadzenie na wiƒôkszym obszarze -- U≈õrednienie z 10 najbli≈ºszymi sƒÖsiadami
kernel_base = np.ones((10,10))
```

```python
# @title
#Wykrywanie krawƒôdzi -- policzenie r√≥≈ºnicy od najbli≈ºszych sƒÖsiad√≥w
kernel_base = np.array(
    [[-1,-1,-1],
     [-1,8,-1],
     [-1,-1,-1]])
kernel = kernel_base
```

```python
# @title
if kernel_base.sum() != 0:
  kernel = kernel_base/kernel_base.sum()
else:
  kernel = kernel_base
```

```python
# @title
kernel
```

```python
# @title
fragment_filtered = filter2D(src=fragment_gray, ddepth=-1, kernel=kernel)
```

```python
# @title
fig = plt.figure()
plt.subplot(232)
plt.imshow(fragment_gray, cmap='gray', vmin=0, vmax=255)
plt.subplot(233)
plt.imshow(fragment_filtered, cmap='gray', vmin=0, vmax=255)
```

```python
# @title
gray_filtered = filter2D(src=gray_mini, ddepth=-1, kernel=kernel)
```

```python
# @title
fig = plt.figure()
plt.subplot(232)
plt.imshow(gray_mini, cmap='gray', vmin=0, vmax=255)
plt.subplot(233)
plt.imshow(gray_filtered, cmap='gray', vmin=0, vmax=255)
```

```python
orig_filtered = filter2D(src=orig_mini, ddepth=-1, kernel=kernel)
```

```python
# @title
fig = plt.figure()
plt.subplot(232)
plt.imshow(orig_mini)
plt.subplot(233)
plt.imshow(orig_filtered)
```
